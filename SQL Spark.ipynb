# https://www.kaggle.com/zygmunt/goodbooks-10k

```python
from pyspark.sql import Row
```

```python
data = spark.sparkContext.textFile('ratings.csv')
data.collect()
```

```python
col_name = data.first()
col_name
```

```python
data = data.filter(lambda x: x!= col_name)
data.collect()
```

```python
def CSV_ROW(row):
    element = row.split(',')
    return Row(book_id=int(element[0]),user_id=int(element[1]),rating=int(element[2]))
```

```python
dt = data.map(CSV_ROW)
dt.collect()
```

```python
df = spark.createDataFrame(dt).cache()
df.createOrReplaceTempView('dt')
```

```python
df.show()
```

```python
df2=df.groupBy('user_id').count().orderBy('count', ascending=False)
```

```python
df2.show()
```

```python
df2.filter(df2['count'] >= 200).count()
```

```python
rating_book = spark.sql('SELECT * FROM dt').groupBy('book_id', 'rating').count().orderBy('book_id', 'rating', ascending=True).show()
```
