```python
import findspark
findspark.init()
findspark.find()
import pyspark
findspark.find()
import sys, re
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('Word Counts')
sc = SparkContext(conf=conf)
```

Tải tệp shakespeare.txt vào một RDD có tên doc:

```python
doc = sc.textFile("shakespeare.txt")
```

Lọc các dòng trống từ RDD, tách các dòng theo khoảng trắng và làm phẳng danh sách các từ thành một danh sách:

```python
flattened = doc.filter(lambda line: len(line) > 0) \
.flatMap(lambda line: re.split('\W+', line))
```

```python
flattened
```

```python
flattened.collect()
```

```python
flattened.take(6)
```

Ánh xạ văn bản thành chữ thường, xóa các chuỗi trống, sau đó chuyển đổi thành khóa / giá trị các cặp trong mẫu (word, 1)

```python
kvpairs = flattened.filter(lambda word: len(word) > 0) \
 .map(lambda word:(word.lower(),1))
```

```python
kvpairs.collect()
```

```python
kvpairs.take(5)
```

Đếm từng từ và sắp xếp kết quả theo thứ tự bảng chữ cái:

```python
countsbyword = kvpairs.reduceByKey(lambda v1, v2: v1 + v2) \
 .sortByKey(ascending=True)
```

```python
countsbyword.collect()
```

```python
countsbyword.take(15)
```

Tìm năm từ được sử dụng nhiều nhất:

```python
# invert the kv pair to make the count the key and sort
topwords = countsbyword.map(lambda x: (x[1],x[0])) \
.sortByKey(ascending=False)
```

```python
topwords.take(5)
```
